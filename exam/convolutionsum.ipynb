```python
import numpy as np
import matplotlib.pyplot as plt

# Take input for x and h as space-separated values
x = list(map(int, input("Enter sequence x (space-separated): ").split()))
h = list(map(int, input("Enter sequence h (space-separated): ").split()))

# Compute the length of the convolution result
l = len(x) + len(h) - 1
y = [0] * l

# Perform linear convolution manually
for i in range(len(x)):
    for j in range(len(h)):
        y[i + j] += x[i] * h[j]

# Print and plot the result
print("Convolution Result:", y)
plt.stem(range(l), y)
plt.xlabel("Index (n)")
plt.ylabel("Amplitude")
plt.title("Linear Convolution of x and h")
plt.grid(True)
plt.show()
     
```
