```python
import numpy as np
import matplotlib.pyplot as plt

# Original discrete signal
x = np.array([1, 3, 1, 2, 3, 4, 1, 1])
t = np.array([-3, -2, -1, 0, 1, 2, 3, 4])  # Time indices

# Delay by 3 units
delay_units = 3
t_delayed = np.arange(t[0], t[-1] + delay_units + 1)  # Extend time index
x_delayed = np.concatenate((np.zeros(delay_units), x))  # Pad with zeros at the start

# Plot original and delayed signals
plt.figure(figsize=(8, 5))

#plt.stem(t, x, linefmt='b-', markerfmt='bo', basefmt='r-', label="Original Signal")
plt.stem(t_delayed, x_delayed)


```

```python

import numpy as np
import matplotlib.pyplot as plt

# Original discrete signal
x = np.array([1, 3, 1, 2, 3, 4, 1, 1])
t = np.array([-3, -2, -1, 0, 1, 2, 3, 4])  # Time indices

# Advance by 3 units (shift left)
unit_advance = 3
t_advanced = np.arange(t[0] - unit_advance, t[-1] - unit_advance + 1 + unit_advance)  # Shift time axis left by 3 and increase length
x_advanced = np.concatenate((x, np.zeros(unit_advance)))  # Keep original values and add 3 zeros at the end

# Plot original and advanced signals
plt.figure(figsize=(8, 5))

plt.stem(t, x, linefmt='b-', markerfmt='bo', basefmt='r-', label="Original Signal")
#plt.stem(t_advanced, x_advanced, linefmt='r-', markerfmt='ro', basefmt='r-', label="Advanced Signal (3 Units)")

plt.xlabel("Time index (n)")
plt.ylabel("Amplitude")
plt.title("Shifting (Advance) of a Discrete Signal")
plt.legend()
plt.grid(True)
plt.show()
```
