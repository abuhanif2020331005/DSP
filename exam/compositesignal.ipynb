```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
fs = 1000    # Sampling frequency in Hz
L = 1000     # Number of samples
t = np.linspace(0, (L-1)/fs, L)  # Time vector

# Define multiple sinusoidal components
f1, A1 = 5, 1.0  # Frequency and amplitude of first sine wave
f2, A2 = 15, 0.5 # Frequency and amplitude of second sine wave
f3, A3 = 30, 0.3 # Frequency and amplitude of third sine wave

# Generate composite signal by summing multiple sine waves
s = A1 * np.sin(2 * np.pi * f1 * t) + A2 * np.sin(2 * np.pi * f2 * t) + A3 * np.sin(2 * np.pi * f3 * t)

# Plot the composite signal
plt.figure(figsize=(8, 4))
plt.plot(t, s, label="Composite Signal", color='b')
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Composite Sinusoidal Wave")
plt.grid(True)
plt.legend()
plt.show()
```
